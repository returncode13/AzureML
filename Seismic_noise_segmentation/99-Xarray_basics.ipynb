```python
import xarray as xr
import numpy as np
```

```python
ds=xr.Dataset(
    {   "amplitudes":(("shots","chans","time"),np.random.rand(3,6,3)) },
    coords={
            "chans":[10,20,30,40,50,60],
            "time":[42,43,44],
            "offset":("chans",list("nnmmff")),       #x=10 and x=20 are nears, 30,40 are mids, 50,60 are fars
            "shots":[0,1,2]
        
    },
    attrs={
        "chans":"number of channels",
        "shots":"shot range",
        "time" :"number of time samples"
    }
)
ds
```

**Shot Domain**  
Selecting a single shot

```python
ds.sel(shots=1).amplitudes.shape
```

**Channel Domain**

```python
ds.sel(chans=[20,30]).amplitudes.shape
```

**Time slices**  
Non-cube**

```python
ds.sel(time=42).amplitudes.shape
```

**Group by** the offsets 

```python
ds.groupby('offset').groups
```

What does the average shot look like? i.e average along the shot direction. i.e. across all shots

```python
ds.amplitudes.mean(dim="shots")
```
