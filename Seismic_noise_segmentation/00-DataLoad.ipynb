```python
import glymur
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=os.path.join(os.getcwd()+"/data")
seq_3=os.path.join(data+'/*1003.jp2')
print(seq_3)
#print(glob.glob(seq_3+"/*.jp2")[0:100])
files=glob.glob(seq_3+"/*.jp2")
files
```

```python
save_location=glob.glob(os.path.join(data+'/*1003.npy'))[0]
for f in files:
    fjp2=glymur.Jp2k(f)
    fname=f.split("/")[-1].split(".")[:2]
    p=""+fname[0]+"."+fname[1]+".npy"
    nmp_f=fjp2[:]
    nmp_f=nmp_f.T
    nf_name=save_location+"/"+p
    print("saving "+f.split("/")[-1]+" to "+p)
    nmp_f.dump(nf_name)
```

```python
plt.imshow(nmp_f.T,cmap="Greys",aspect="auto")
plt.tight_layout()
plt.show()
```
